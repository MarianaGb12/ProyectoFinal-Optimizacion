```python
from sklearn.model_selection import StratifiedKFold, cross_val_predict, train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score, accuracy_score
from imblearn.over_sampling import SMOTE

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df_train[inputs], df_train[outputs], test_size=0.2, random_state=42, stratify=df_train[outputs])

# Apply SMOTE to the training data
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

# Define Stratified K-Fold
skf = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

# Define the Random Forest Model
rf_model = RandomForestClassifier(**best_params)

# Perform Cross-Validation and get predictions
y_pred = cross_val_predict(rf_model, X_train_resampled, y_train_resampled, cv=skf, n_jobs=-1)

# Get the true values from the dataset (y_train_resampled)
y_true = y_train_resampled

# Compute confusion matrix
cm = confusion_matrix(y_true, y_pred)
print("Confusion Matrix:\n", cm)

# Compute accuracy
accuracy = accuracy_score(y_true, y_pred)
print("Accuracy:", accuracy)

# Compute precision
precision = precision_score(y_true, y_pred, average='macro')  # or 'micro'/'weighted'
print("Precision:", precision)

# Compute recall
recall = recall_score(y_true, y_pred, average='macro')  # or 'micro'/'weighted'
print("Recall:", recall)

# Compute F1 score
f1 = f1_score(y_true, y_pred, average='macro')  # or 'micro'/'weighted'
print("F1 Score:", f1)
```

4. **Recall (0.825):**
   - Recall indicates the proportion of actual positives that were correctly identified. A recall of 82.5% suggests that the model is fairly good at minimizing false negatives.

5. **F1 Score (0.825):**
   - The F1 Score is the harmonic mean of precision and recall. An F1 score of 82.5% indicates a good balance between precision and recall.
